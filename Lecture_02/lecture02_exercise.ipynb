Write two Python functions to find the minimum number in a list. The first function should compare each number to every other number in the list, with a computational cost of $\mathcal{O}(n^2)$. The second function should be linear, i.e. have computational cost of $\mathcal{O}(n)$.

Verify the costs of the two algorithms empirically using the `timeit.timeit` function.
